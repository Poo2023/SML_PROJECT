```python
from os import listdir
from os.path import join
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.utils import to_categorical, Sequence
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, Embedding, LSTM, concatenate, Input, RepeatVector
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.callbacks import ModelCheckpoint
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Enable GPU for faster training if available
print("GPU Available:", tf.config.list_physical_devices('GPU'))

# Setting up directory paths
dir_name = '/content/pix2code/datasets/web/training_features'

# Read a file and return a string
def load_doc(filename):
    with open(filename, 'r') as file:
        text = file.read()
    return text

# Load data and organize it into images and text lists
def load_data(data_dir):
    text = []
    images = []
    all_filenames = listdir(data_dir)
    all_filenames.sort()
    for filename in all_filenames:
        if filename.endswith(".npz"):
            image = np.load(join(data_dir, filename))
            images.append(image['features'])
        else:
            syntax = '<START> ' + load_doc(join(data_dir, filename)) + ' <END>'
            syntax = ' '.join(syntax.split())
            syntax = syntax.replace(',', ' ,')
            text.append(syntax)
    images = np.array(images, dtype=np.float32)
    return images, text

train_features, texts = load_data(dir_name)

# Initialize tokenizer
tokenizer = Tokenizer(filters='', split=" ", lower=False)
tokenizer.fit_on_texts([load_doc('/content/Screenshot-to-code/Bootstrap/resources/bootstrap.vocab')])

# Vocabulary settings
vocab_size = len(tokenizer.word_index) + 1
train_sequences = tokenizer.texts_to_sequences(texts)
max_sequence = max(len(s) for s in train_sequences)
max_length = 48

# Define a custom data generator
class DataGenerator(Sequence):
    def __init__(self, sequences, features, batch_size, max_sequence, max_length, vocab_size):
        self.sequences = sequences
        self.features = features
        self.batch_size = batch_size
        self.max_sequence = max_sequence  # Maximum sequence length for padding
        self.max_length = max_length      # Fixed length of 48 tokens for model input
        self.vocab_size = vocab_size
        self.indices = np.arange(len(sequences))

    def __len__(self):
        # Number of batches per epoch
        return int(np.ceil(len(self.sequences) / self.batch_size))

    def __getitem__(self, index):
        # Generate one batch of data
        batch_indices = self.indices[index * self.batch_size:(index + 1) * self.batch_size]
        X, y, image_data = [], [], []

        for idx in batch_indices:
            seq = self.sequences[idx]
            for i in range(1, len(seq)):
                # Step 1: Create input and output sequences
                in_seq, out_seq = seq[:i], seq[i]

                # Step 2: Pad in_seq to max_sequence length
                in_seq = pad_sequences([in_seq], maxlen=self.max_sequence, dtype='int32')[0]

                # Step 3: Truncate to the last 48 tokens to ensure consistent input length
                in_seq = in_seq[-self.max_length:]

                # Step 4: One-hot encode the output token
                out_seq = to_categorical([out_seq], num_classes=self.vocab_size)[0].astype(np.float32)

                # Step 5: Ensure the length of in_seq is exactly 48
                if len(in_seq) < self.max_length:
                    in_seq = np.pad(in_seq, (self.max_length - len(in_seq), 0), 'constant')

                # Append data for the batch
                image_data.append(self.features[idx])  # Image features as float32
                X.append(in_seq)                       # Input sequence with exactly 48 tokens (int32)
                y.append(out_seq)                      # One-hot encoded output token (float32)

        # Convert lists to numpy arrays with the correct shape and dtype
        return (np.array(image_data, dtype=np.float32), np.array(X, dtype=np.int32)), np.array(y, dtype=np.float32)

# Split data for training and validation (e.g., 90% train, 10% validation)
split_index = int(0.9 * len(train_sequences))
train_sequences_data = train_sequences[:split_index]
val_sequences_data = train_sequences[split_index:]

train_features_data = train_features[:split_index]
val_features_data = train_features[split_index:]

# Set up parameters and create training and validation generators
batch_size = 8
max_sequence = max(len(s) for s in train_sequences)
max_length = 48

train_generator = DataGenerator(train_sequences_data, train_features_data, batch_size, max_sequence, max_length, vocab_size)
val_generator = DataGenerator(val_sequences_data, val_features_data, batch_size, max_sequence, max_length, vocab_size)

# Image encoder model
image_model = Sequential([
    Conv2D(16, (3, 3), padding='valid', activation='relu', input_shape=(256, 256, 3)),
    Conv2D(16, (3, 3), activation='relu', padding='same', strides=2),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    Conv2D(32, (3, 3), activation='relu', padding='same', strides=2),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    Conv2D(64, (3, 3), activation='relu', padding='same', strides=2),
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    Flatten(),
    Dense(1024, activation='relu'),
    Dropout(0.3),
    Dense(1024, activation='relu'),
    Dropout(0.3),
    RepeatVector(max_length)
])

visual_input = Input(shape=(256, 256, 3))
encoded_image = image_model(visual_input)

# Language model setup (without mask_zero=True)
language_input = Input(shape=(max_length,))
language_model = Embedding(vocab_size, 50, input_length=max_length)(language_input)
language_model = LSTM(128, return_sequences=True)(language_model)
language_model = LSTM(128, return_sequences=True)(language_model)

# Decoder setup
decoder = concatenate([encoded_image, language_model])
decoder = LSTM(512, return_sequences=True)(decoder)
decoder = LSTM(512, return_sequences=False)(decoder)
decoder = Dense(vocab_size, activation='softmax')(decoder)

# Compile the model
model = Model(inputs=[visual_input, language_input], outputs=decoder)
optimizer = RMSprop(learning_rate=0.0001, clipvalue=1.0)
model.compile(loss='categorical_crossentropy', optimizer=optimizer)

# Checkpoint configuration: save only the best model
filepath = "org-weights-best-loss-{loss:.4f}.weights.h5"
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True, save_weights_only=True)
callbacks_list = [checkpoint]

# Train the model using the data generators and capture history
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    validation_data=val_generator,
    validation_steps=min(10, len(val_generator)),  # Adjust based on actual validation data size
    epochs=200,
    callbacks=callbacks_list,
    verbose=1
)

# Plot training and validation loss
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()
```

```python
from nltk.translate.bleu_score import corpus_bleu
import numpy as np

# Load the `eval_light` data using the same `load_data` function
eval_features, eval_texts = load_data('/content/pix2code/datasets/web/eval_features/')  # Replace with actual eval_light path

# Modified generate_desc with Beam Search for better predictions
def generate_desc_with_beam_search(model, tokenizer, photo, max_length, beam_width=3):
    start_token = '<START>'
    end_token = '<END>'
    sequences = [[start_token, 1.0]]  # Initialize with start token and probability of 1

    for _ in range(max_length*2):
        all_candidates = []
        for seq, score in sequences:
            sequence = tokenizer.texts_to_sequences([seq])[0]
            sequence = pad_sequences([sequence], maxlen=max_length)
            yhat = model.predict([photo.reshape(1, *photo.shape), sequence], verbose=0)
            top_k_indices = np.argsort(yhat[0])[-beam_width:]  # Get indices of top k tokens

            for idx in top_k_indices:
                word = tokenizer.index_word.get(idx)
                if word is None:
                    continue
                new_seq = seq + ' ' + word
                new_score = score * yhat[0][idx]
                candidate = [new_seq, new_score]
                all_candidates.append(candidate)

        # Select top `beam_width` sequences
        sequences = sorted(all_candidates, key=lambda x: x[1], reverse=True)[:beam_width]

        # Check if any sequence has generated the `<END>` token
        if any(end_token in seq for seq, _ in sequences):
            break

    # Return the sequence with the highest probability
    best_sequence = max(sequences, key=lambda x: x[1])[0]
    return best_sequence

# Function to evaluate the model, print predicted and real data, and calculate BLEU score
def evaluate_model_with_generator(model, tokenizer, eval_features, eval_texts, max_length):
    actual, predicted = [], []
    for i, photo in enumerate(eval_features):
        # Generate a description for the evaluation image using beam search
        yhat = generate_desc_with_beam_search(model, tokenizer, photo, max_length)

        # Use ground-truth text from `eval_texts`
        actual_text = eval_texts[i]

        # Add the actual and predicted descriptions to lists for BLEU calculation
        actual.append([actual_text.split()])
        predicted.append(yhat.split())

        # Print the real and predicted descriptions
        print(f"Image {i + 1}:")
        print(f"Ground-Truth Description: {actual_text}")
        print(f"Generated Description: {yhat}\n")

    # Calculate BLEU score
    bleu = corpus_bleu(actual, predicted)
    print(f"Final BLEU score: {bleu}")
    return bleu

# Run the evaluation with the modified generate_desc using beam search
bleu_score = evaluate_model_with_generator(model, tokenizer, eval_features, eval_texts, max_length)
print(f"Final BLEU score: {bleu_score}")
```

```python
from os import listdir
from os.path import join
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
from nltk.translate.bleu_score import corpus_bleu
# from compiler.classes.Compiler import Compiler
import sys
sys.path.append('/content/Screenshot-to-code/Bootstrap/compiler')
from classes.Compiler import Compiler
# Setting up directory paths
dir_name = '/content/Screenshot-to-code/Bootstrap/resources/eval_light'

# Load data and organize it into images and text lists
def load_data(data_dir):
    text = []
    images = []
    all_filenames = listdir(data_dir)
    all_filenames.sort()
    for filename in all_filenames:
        if filename.endswith(".npz"):
            image = np.load(join(data_dir, filename))
            images.append(image['features'])
        else:
            syntax = '<START> ' + load_doc(join(data_dir, filename)) + ' <END>'
            syntax = ' '.join(syntax.split())
            syntax = syntax.replace(',', ' ,')
            text.append(syntax)
    images = np.array(images, dtype=np.float32)
    return images, text

def load_doc(filename):
    with open(filename, 'r') as file:
        text = file.read()
    return text

# Load evaluation data
eval_features, eval_texts = load_data(dir_name)

# Initialize tokenizer
tokenizer = Tokenizer(filters='', split=" ", lower=False)
tokenizer.fit_on_texts([load_doc('/content/Screenshot-to-code/Bootstrap/resources/bootstrap.vocab')])

# Vocabulary and maximum length
vocab_size = len(tokenizer.word_index) + 1
max_length = 48

# Function to map an integer to a word
def word_for_id(integer, tokenizer):
    for word, index in tokenizer.word_index.items():
        if index == integer:
            return word
    return None

# Generate a description for an image
def generate_desc(model, tokenizer, photo, max_length):
    photo = np.array([photo])
    in_text = '<START> '
    print('\nPrediction:\n<START> ', end='')
    for _ in range(150):
        sequence = tokenizer.texts_to_sequences([in_text])[0]
        sequence = pad_sequences([sequence], maxlen=max_length)
        yhat = model.predict([photo, sequence], verbose=0)
        yhat = np.argmax(yhat)
        word = word_for_id(yhat, tokenizer)
        if word is None:
            break
        in_text += word + ' '
        print(word + ' ', end='')
        if word == '<END>':
            break
    return in_text

# Evaluate the model using BLEU score
def evaluate_model(model, descriptions, photos, tokenizer, max_length):
    actual, predicted = list(), list()
    for i in range(len(descriptions)):
        yhat = generate_desc(model, tokenizer, photos[i], max_length)
        print('\n\nReal:\n' + descriptions[i])
        actual.append([descriptions[i].split()])
        predicted.append(yhat.split())
    bleu = corpus_bleu(actual, predicted)
    return bleu, actual, predicted

# Evaluate the trained model
bleu, actual, predicted = evaluate_model(model, eval_texts, eval_features, tokenizer, max_length)

# Compile tokens into HTML and CSS
dsl_path = "/content/Screenshot-to-code/compiler/assets/web-dsl-mapping.json"
compiler = Compiler(dsl_path)
compiled_website = compiler.compile(predicted[0], 'index.html')
print("\nCompiled HTML:", compiled_website)
print("BLEU Score:", bleu)
```
